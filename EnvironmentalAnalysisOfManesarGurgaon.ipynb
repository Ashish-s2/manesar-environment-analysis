#  Environmental Analysis of Manesar, Gurgaon
###  Project | Data Analytics (2025)

This project analyzes the environmental and socio-economic conditions of Manesar, a fast-developing industrial region in Gurgaon, Haryana.  
The aim is to study multiple factors affecting the region including climate, pollution, land use, water resources, biodiversity, and risk management.

**Tools used:** Python, Pandas, Seaborn, Matplotlib  
**Data Sources:** OpenAQ, CPCB, Census India, ISRO Bhuvan, Global Forest Watch, Meteostat, etc.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
socio = pd.read_csv("manesar_socio_economic_real.csv")
climate = pd.read_csv("manesar_climate_conditions.csv")
geo = pd.read_csv("manesar_geological_topographical.csv")
forest = pd.read_csv("manesar_forest_biodiversity.csv")
water = pd.read_csv("manesar_water_resources.csv", encoding='ISO-8859-1')
land = pd.read_csv("manesar_land_use_urbanization.csv")
disaster = pd.read_csv("manesar_disaster_risk.csv",  encoding='ISO-8859-1')
pollution = pd.read_csv("manesar_pollution_environment.csv" , encoding='ISO-8859-1')
```

##  Climate Conditions in Manesar
We analyze monthly temperature and rainfall patterns from 2022â€“2024.


```python
climate['Date'] = climate['Year'].astype(str) + '-' + climate['Month']
climate['Date'] = pd.to_datetime(climate['Date'], format='%Y-%b')

plt.figure(figsize=(12,6))
sns.lineplot(data=climate, x='Date', y='Avg Temperature (Â°C)', marker='o', color='crimson')
plt.title("Average Monthly Temperature in Manesar (2022â€“2024)")
plt.xlabel("Date")
plt.ylabel("Temperature (Â°C)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(data=climate, x='Date', y='Total Rainfall (mm)', color='royalblue')
plt.title("Monthly Rainfall in Manesar (2022â€“2024)")
plt.xlabel("Date")
plt.ylabel("Rainfall (mm)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

- Peak rainfall observed between Juneâ€“August, indicating monsoon impact.
- Temperatures peak in Mayâ€“June (~33â€“35Â°C), drop in winter to ~14Â°C.

## ðŸ§± Geology & Terrain
This section examines elevation, slope, and soil types across Manesar.


```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=geo, x="Elevation (m)", y="Slope (%)", hue="Soil Type", s=100)
plt.title("Elevation vs Slope in Manesar")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=geo, x="Soil Type", hue="Soil Type", palette="viridis", legend=False)
plt.title("Soil Types in Manesar")
plt.tight_layout()
plt.show()

```

- Elevation ranges between 269â€“275 m, slope gentle (1â€“2%).
- Dominant soil: Sandy Loam â€” good for moderate agriculture and construction.


##  Water Resources in Manesar
We examine groundwater level decline and water quality.


```python
plt.figure(figsize=(10,5))
sns.lineplot(data=water, x='Year', y='Groundwater Level (m below ground)', marker='o', color='teal')
plt.title("Groundwater Level in Manesar")
plt.ylabel("Depth (m)")
plt.grid(True)
plt.show()

plt.figure(figsize=(8,4))
sns.barplot(data=water, x='Year', y='Water Quality Index (0-100)', color='skyblue')
plt.title("Water Quality Index Over Time")
plt.ylim(50, 80)
plt.tight_layout()
plt.show()
```

- Groundwater dropping from ~23m (2018) to ~26m (2023) â†’ depletion risk.
- WQI is borderline safe (~60â€“70); better monitoring needed.


##  Forest & Biodiversity
Forest area vs tree cover loss (2018â€“2023).


```python
plt.figure(figsize=(10,5))
sns.lineplot(data=forest, x='Year', y='Recorded Forest Area (sq km)', label='Forest Area')
sns.lineplot(data=forest, x='Year', y='Tree Cover Loss (ha)', label='Tree Loss', color='red')
plt.title("Forest Area and Tree Loss")
plt.grid(True)
plt.legend()
plt.show()
```

- Forest area shrinking yearly; consistent tree cover loss observed.
- Biodiversity richness index dropped slightly â†’ needs conservation.


##  Land Use and Urban Growth
We analyze changes in built-up, agriculture and open spaces.


```python
plt.figure(figsize=(10,5))
sns.lineplot(data=land, x='Year', y='Built-up Area (sq km)', label='Built-up')
sns.lineplot(data=land, x='Year', y='Agricultural Land (sq km)', label='Agriculture')
sns.lineplot(data=land, x='Year', y='Green/Open Spaces (sq km)', label='Open Spaces')
plt.title("Land Use Changes in Manesar")
plt.ylabel("Area (sq km)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

- Urban areas increased by 130% (2010â€“2023).
- Green and agri land declining â†’ ecological stress.


##  Pollution & Waste
We analyze PM2.5 levels and waste generation trends.


```python
plt.figure(figsize=(8,4))
sns.barplot(data=pollution, x='Year', y='Avg PM2.5 (Âµg/mÂ³)', color='salmon')
plt.title("Air Pollution: PM2.5 in Manesar")
plt.tight_layout()
plt.show()

plt.figure(figsize=(8,4))
sns.lineplot(data=pollution, x='Year', y='Waste Generation (tons/day)', marker='o', color='gray')
plt.title("Waste Generation Trend")
plt.grid(True)
plt.tight_layout()
plt.show()
```

- PM2.5 > 100 Âµg/mÂ³ in most years â†’ Very Poor air quality.
- Waste generation increasing steadily with urban growth.


##  Disaster Risk in Manesar
Floods, earthquakes, and heatwaves evaluated.


```python
plt.figure(figsize=(7,5))
sns.barplot(data=disaster, x='Risk Type', y='Severity (Scale 1-5)', hue='Risk Type', palette='flare', legend=False)
plt.title("Disaster Risk Severity in Manesar")
plt.tight_layout()
plt.show()
```

- Earthquake (Zone IV) and heatwaves show high severity.
- Floods low-moderate due to urban drainage.
